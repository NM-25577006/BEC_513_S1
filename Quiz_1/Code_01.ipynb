```python
#!/usr/bin/env python3
# Usage: cat data.tsv | python select_lines.py <query_file> <column_number>
# Example: zcat q1_data.tsv.gz | python select_lines.py to_select.tsv 2

from sys import stdin, stdout, stderr, argv


query_file = argv[1]
col_num = int(argv[2]) - 1  # Convert to zero-based index

# Read query patterns into a set for O(1) lookup
with open(query_file, 'r') as qf:
    query_patterns = {line.strip() for line in qf if line.strip()}

print(f"Loaded {len(query_patterns)} patterns from {query_file}", file=stderr)
print(f"Filtering column {argv[2]} (0-indexed: {col_num})", file=stderr)

matches = 0
for line in stdin:
    fields = line.rstrip('\n').split('\t')

    # Check if column exists
    if len(fields) <= col_num:
        continue

    # Check if value matches any query pattern
    if fields[col_num] in query_patterns:
        stdout.write(line)
        matches += 1

print(f"Found {matches} matching lines", file=stderr)
```
